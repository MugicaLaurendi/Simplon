### Data Cleaning

```python
dummies_train_lot_size_units = pd.get_dummies(data_train["lot_size_units"])
dummies_test_lot_size_units = pd.get_dummies(data_test["lot_size_units"])

data_train["lot_size"] = data_train["lot_size"] * dummies_train_lot_size_units["acre"] + data_train["lot_size"] * dummies_train_lot_size_units["sqft"] * 43560
data_test["lot_size"] = data_test["lot_size"] * dummies_test_lot_size_units["acre"] + data_test["lot_size"] * dummies_test_lot_size_units["sqft"] * 43560

data_train
```

```python
data_train = data_train.dropna()
data_test = data_test.dropna()
```

```python
data_train = data_train.drop(columns = "lot_size_units")
data_train = data_train.drop(columns = "size_units")

data_test = data_test.drop(columns = "lot_size_units")
data_test = data_test.drop(columns = "size_units")
```

```python
data_train["zip_code"] = data_train["zip_code"].apply(lambda x : str(x))
data_test["zip_code"] = data_test["zip_code"].apply(lambda x : str(x))
```
