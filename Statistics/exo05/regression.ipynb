```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy.stats import norm
import matplotlib.lines as lines
import statsmodels.api as sm

df_tot = pd.read_csv('akinator.csv')
df_tot
```

```python
df = {'Taille':df_tot['Taille (cm)'], 'Taille de pied':df_tot['Taille de pied']}
df = pd.DataFrame(data=df)
df
```

```python
y_mes = np.array(df['Taille de pied'])
x = np.array(df['Taille'])

y1 = np.array(df['y1'])
y2 = np.array(df['y2'])
```

```python
def r2(y_meas,y_pred) :
    y_moy = np.mean(y_meas)
    return 1 - (sum((y_meas - y_pred)**2)/sum((y_mes - y_moy)**2))
```

```python
r2_y1 = r2(y_mes, y1)
r2_y2 = r2(y_mes, y2)
print(r2_y1)
print(r2_y2)
```
