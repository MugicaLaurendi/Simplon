# Exploration des données
- Télécharger le fichier au format CSV
- Ouvrir le fichier dans un notebook et le convertir en un dataframe
- Tracer la distribution des lancés pour chacun des groupes (avec des histogrammes de matplotlib). Pour ce faire il faudra convertir les mesures rentré en "string" au format "float".
- Tracer la distribution de l'ensemble des lancés (population totale) (avec des histogrammes de matplotlib). Pour ce faire il faudra restructurer la donnée en une seule colonne (la méthode "reshape" peut être une approche)

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('akinator IA3 - Feuille 3.csv')
```

```python
# Nettoyage des données

df = df.drop(df.columns[0], axis=1)

header = df.head()

def cleaner(x):
    if type(x) in (int, float): return x
    if ',' in x: x = x.replace(',', '.')
    return float(x)
    

for column in header: df[column] = df[column].apply(cleaner)
```

```python
# Tracé des histogrammes par groupes

for column in header:
    plt.hist(df[column])
    plt.title(column)
    plt.xlabel('Distance (cm)')
    plt.show()

for column in header:
    plt.hist(df[column])
    plt.title('Groupes superposés')
    plt.xlabel('Distance (cm)')
```

```python
# Tracé d'un histogramme pour les valeurs de tous les groupes

valeurs = []

for column in header: valeurs += df[column].values.tolist()

plt.hist(valeurs, bins=21)
plt.xlabel('Distance (cm)')
plt.title('Tous les groupes')
plt.show()
```
