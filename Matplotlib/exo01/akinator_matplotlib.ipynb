```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


csv = "akinator.csv"
df = pd.read_csv(csv)


def sucre_sale(x) :
    if x == 'Sal√©' or x == 'sal√©' :
        return 'sal√©'
    if x == 'Sucr√©' or x == 'sucr√©' :
        return 'sucr√©'
    return 'both'

df['Sucr√© ou sal√©?'] = df['Sucr√© ou sal√©?'].apply(sucre_sale)


df.groupby('Sucr√© ou sal√©?')['Nom'].nunique().plot(kind='bar')
plt.show()
```

```python
ax = plt.gca()

df.plot(kind='line',x='Nom',y='Taille (cm)',ax=ax)
df.plot(kind='line',x='Nom',y='Taille de pied',color = 'red',ax=ax,figsize=[20, 8], rot=80,)

plt.show()
```

```python
fig, ax = plt.subplots()

fruits = df['Sucr√© ou sal√©?']

bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

ax.bar(fruits, color=bar_colors)

ax.set_ylabel('fruit supply')
ax.set_title('Fruit supply by kind and color')
ax.legend(title='Fruit color')

plt.show()
```
