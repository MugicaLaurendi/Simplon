### Création d'un DataFrame

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
df = pd.DataFrame({"Prénom" : ["Django", "Louise", "Laurendi", "Rémy", "Nano", "Ake G. Prince"],
                   "Age" : [18, 22, 25, 27, 29, 32],
                  "Date à laquelle j'ai cesser de croire en l'existence du Père Noël": ["24/12/2012", "19/09/2004" ,"24/12/2005", "11/09/2001",
                                                                                        "31/12/1997", "12/12/1995"],
                   "Terre plate ?" : ["Non", "Oui", "Non", "Oui", "Oui", "Oui"],
                  "Terre ronde ?": ["Oui", "Non", "Oui", "Non", "Non", "Non"],
                  "Terre ellipsoïdale ?" : [np.nan, np.nan, np.nan, np.nan, "Oui", np.nan]},
                 index = [1,2,3,4,5,6])

pd.set_option("display.max.columns", None)
df
```

### Création d'un 2ème DataFrame

```python
df2 = pd.DataFrame({"Longueur des cheveux" : ["9,8",27.5,"12,2","4,8","51,3", np.nan]},
                  index = [1,2,3,4,5,6])
df2
```

### Fusion des 2 DataFrames

```python
df = pd.concat([df, df2], axis = 1)
df
```

### Afficher les informations du DataFrame

```python
df.info()
df.nunique()
df.isna().sum()
df.count() / df.shape[0] *100
```

### Retrait d'une colonne redondante ou avec trop de valeurs manquantes:

```python
df = df.drop(columns = ["Terre ellipsoïdale ?"])
del df["Terre ronde ?"]
df
```

### Renommer une colonne :

```python
df.rename(columns={"Date à laquelle j'ai cesser de croire en l'existence du Père Noël" : "DLJCCEPN"}, inplace = True)
df
```

### Transformer une colonne en Date

```python
df["DLJCCEPN"] = pd.to_datetime(df["DLJCCEPN"], infer_datetime_format=True)
df
```

### convertir des valeurs 

```python
df["Longueur des cheveux"] = df["Longueur des cheveux"].replace("," , ".", regex = True).astype(float)  # Nano Solution (The Best)

# Or

df["Longueur des cheveux"] = df["Longueur des cheveux"].astype(str)
df["Longueur des cheveux"] = df["Longueur des cheveux"].apply(lambda x: x.replace(",", "."))
df["Longueur des cheveux"] = df["Longueur des cheveux"].astype(float)
# df["Longueur des cheveux"] = df["Longueur des cheveux"].astype(str).apply(lambda x: x.replace(",", ".")).astype(float) Mickael & Camille 🤫 <3
```

```python
df["Longueur des cheveux"] = df["Longueur des cheveux"].fillna(0)
df
```

### GroupBy


```python
df.groupby("Terre plate ?")["Longueur des cheveux"].mean().plot(kind = "bar", ylabel = "Longueur des cheveux (cm)", rot= 0);
```

### Get_dummies


```python
dummies = pd.get_dummies(df["Terre plate ?"])
df = df.join(dummies)
df["Non"].sum()
df
```

```python
sns.heatmap(df.corr(), annot = True, linewidth = .5);
```

```python
df["Age AJCCEPN"] = 2022 - df["Age"]
df['year'] = pd.to_datetime(df['DLJCCEPN']).dt.strftime('%Y').astype(int)
df["Age AJCCEPN"] = df["year"] - df["Age AJCCEPN"]
df
```

```python
df.groupby("Terre plate ?")["Age AJCCEPN"].mean().plot(kind= "barh" );
```
